```python
import pandas as pd
import altair as alt
import app_wrangling as aw
import app_graphing as ag
import numpy as np
```

```python
boardgame_data = pd.read_csv(
    "../../data/app_data/board_game.csv", parse_dates=["year_published"], index_col=0
)

boardgame_data["year_published"] = pd.to_datetime(
    boardgame_data["year_published"], format="%Y"
)

# convert NA values for these features to a value
values = {"category": "Unknown", "mechanic": "Unknown", "publisher": "Unknown"}
boardgame_data.fillna(value=values, inplace=True)

# create lists from strings
cats_split = ["category", "mechanic", "publisher"]
boardgame_data[cats_split] = (
    boardgame_data[cats_split].stack().str.split(r",(?![+ ])").unstack()
)
```

```python
data=boardgame_data
col="mechanic"
year_in=2005
year_out=2010
color_="#ff7f0e"

plot_check = aw.call_boardgame_top(data, col, year_in, year_out)

list_ = plot_check[col].to_list()

plot_data=data[aw.call_bool_series_or(data, col, list_)]
```

```python
plot_first = aw.call_boardgame_radio(data, col, ['Crayon Rail System'])
#plot_first = aw.call_boardgame_top_density(data, col, 1900, 2200, 0)
```

```python
test = aw.bin_rating(plot_first)
aw.density_transform(test, col)
```

```python
plot_mean = plot_first.explode('group').groupby('group')['average_rating'].mean().to_frame('average_rating').reset_index()

chart2 = aw.density_transform(plot_first, 'mechanic')
```

```python
check = list(np.arange(0,10.5,0.5))
set_ = list(np.arange(0.5,10.5,0.5))
plot_first['average_rating']=pd.cut(plot_first['average_rating'], bins=check, labels=set_)
plot_first
```

```python
plot_density = plot_first.explode('group').groupby(["average_rating", "group"])['category'].count().to_frame('density').reset_index()

names = list(plot_density['group'].unique())
chart = []

for x in names:
     temp = plot_density[plot_density['group'] == x]
     temp['density'] = temp['density']/temp['density'].sum()
     temp['mean'] = plot_mean[plot_mean['group']==x]['average_rating']
     chart.append(temp)

chart = pd.concat(chart).reset_index().drop(columns='index')

```
